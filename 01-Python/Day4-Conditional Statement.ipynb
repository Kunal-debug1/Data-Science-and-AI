```python
#24. Write a program to determine whether a triangle is equilateral, isosceles, or scalene.
# Input sides of the triangle
side1 = float(input("Enter length of first side: "))
side2 = float(input("Enter length of second side: "))
side3 = float(input("Enter length of third side: "))

# First, check if the sides can actually form a triangle (triangle inequality)
if (side1 + side2 > side3) and (side1 + side3 > side2) and (side2 + side3 > side1):
    if side1 == side2 == side3:
        print("The triangle is equilateral.")
    elif side1 == side2 or side1 == side3 or side2 == side3:
        print("The triangle is isosceles.")
    else:
        print("The triangle is scalene.")
else:
    print("The given sides do not form a valid triangle.")
```

```python
#43. Convert a Roman numeral to an integer 
def roman_to_int(s):   
    roman = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}   
    total = 0   
    for i in range(len(s)):   
        if i > 0 and roman[s[i]] > roman[s[i - 1]]:   
            total += roman[s[i]] - 2 * roman[s[i - 1]]   
        else:   
            total += roman[s[i]]   
    return total   
 
num = input("Enter Roman numeral: ").upper()   
print("Integer:", roman_to_int(num))
```

```python
#50. Find the day of the week for a given date (without built-in functions) 
def day_of_week(d, m, y):   
    if m < 3:   
        m += 12   
        y -= 1   
    k = y % 100   
    j = y // 100   
    day = (d + (13 * (m + 1)) // 5 + k + (k // 4) + (j // 4) - 2 * j) % 7   
    days = ["Saturday", "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]   
    return days[day]   
 
day, month, year = map(int, input("Enter date (DD MM YYYY): ").split())   
 
print("Day of the Week:", day_of_week(day, month, year))  
```
