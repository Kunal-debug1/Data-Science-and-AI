# 🌐 Scope in Python :

📍 What is Scope?

Scope defines which variables are accessible at which parts of the program.

🔸 Local Scope:

Variable declared inside a function. Only accessible within that function.

```python
def example():
    x = 10  # Local variable
    print(x)
```

🔸 Global Scope:

Declared outside all functions. Can be accessed anywhere in the code.


```python
x = 50  # Global variable

def show():
    print(x)
```

🔸 global Keyword:  – To modify global variables inside a function

Use inside function to modify a global variable.

```python
x = 5
def modify():
    global x
    x += 1
```

🔸 nonlocal Keyword:

Used inside nested functions to modify a variable from the enclosing (non-global) scope.

```python
def outer():
    x = 10
    def inner():
        nonlocal x
        x += 1
    inner()
    return x  # Returns 11
```

# 🎯 Function Parameters and Arguments

✅1. Positional Arguments (Order matters) :

    Passed in order of parameters.

```python
def info(name, age):
    print(name, age)

info("Alice", 25)
```

✅2. Keyword Arguments (Specify using key=value)
   
     Specify the parameter name during the call.

```python
info(age=22, name="Alice")
```

✅3. Default Arguments

     Assign a default value to a parameter.

```python
def greet(name="Guest"):
    print("Hello", name)

greet()        # Hello Guest
greet("John")  # Hello John
```

✅4. Variable-Length Arguments
   
    *args – Multiple positional arguments (tuple)/Collects extra positional arguments as a tuple.

```python
def add_all(*args):
    return sum(args)

add_all(1, 2, 3, 4)  # Returns 10
```

✅ **kwargs – Multiple keyword arguments (dictionary)
 
    Collects extra keyword arguments as a dictionary.

```python
def show_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

show_info(name="Alice", age=25)
```

# 🔁 Recursion

When a function calls itself.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

📌 Key Components of Recursion:
- Base Case – when to stop recursion

- Recursive Case – calls itself

⚠️ Be careful with stack overflow due to too many recursive calls.

✅ Tail Recursion (Optimized Recursion):

    Some languages optimize tail-recursive functions — Python doesn't natively

# 🧠 PART 2: MODULES IN PYTHON
✅ What is a Module?

A module is a file containing Python code (.py) — functions, variables, classes — to be reused.

🔹 Using Built-in Modules

```python
import math

print(math.sqrt(16))  # Output: 4.0
```

🔹 Importing Techniques
| Syntax                    | Use                            |
| ------------------------- | ------------------------------ |
| `import module`           | Imports the whole module       |
| `from module import func` | Imports only specific function |
| `import module as alias`  | Gives a shortcut name          |

Examples:

```python
import random
print(random.randint(1, 10))

from math import pi
print(pi)

import datetime as dt
print(dt.datetime.now())
```
