# ✨ 4. Writing Clean & Maintainable Code
📖 Follow PEP8 – Python's Style Guide
| Rule/Tip                            | Example                     |
| ----------------------------------- | --------------------------- |
| Use 4 spaces for indentation        | ✅ `if True:` + 4 spaces     |
| Limit lines to 79 characters        | Wrap long lines properly    |
| Add docstrings to functions/classes | Use `""" triple quotes """` |
| Use `snake_case` for variables      | `user_name`, `total_count`  |
| Keep functions short & reusable     | One function = one task     |
| Use meaningful variable names       | ❌ `x = 5`, ✅ `age = 5`      |
| Add blank lines between functions   | Improves readability        |


🧰 Tools to Enforce Clean Code
- flake8 – Code linting (PEP8)

- black – Auto formatting

- isort – Sort and organize imports

- .editorconfig – Unified code styles across editors
