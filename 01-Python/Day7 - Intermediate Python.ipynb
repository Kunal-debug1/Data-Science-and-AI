# 🔷 File Handling in Python
📘 Why File Handling?

Programs often need to store and retrieve data permanently, which isn't possible with variables alone (since data in variables disappears after program execution). File handling allows:

- Reading configuration or data files
- Writing logs
- Saving user data, models, or results

✅ open() Function — The Entry Point

```python
file_object = open("Roadmap.pdf", "mode")
```

- Returns a file object to interact with the file.
- Needs a filename and a mode (string).

📂 File Modes Explained :
| Mode  | Description                     | Notes                                |
| ----- | ------------------------------- | ------------------------------------ |
| `'r'` | Read (default)                  | File must exist                      |
| `'w'` | Write (overwrite or create)     | Destroys existing data               |
| `'a'` | Append                          | Data is added to the end             |
| `'x'` | Create file, error if it exists | Raises `FileExistsError`             |
| `'b'` | Binary mode                     | Use with images, videos, etc.        |
| `'+'` | Update mode (read/write)        | Combine with other modes like `'r+'` |

📌 Examples:

```python
open("data.txt", "rb")  # read binary
open("log.txt", "a+")   # append + read
```

# 🔷 Error & Exception Handling
📘 Why Exception Handling?

Programs can crash due to unexpected events (e.g., file not found, divide by zero). Exception handling allows you to catch such errors and recover gracefully without crashing the program.

✅ The try-except Block :

✅ Handling Specific Exceptions :

try:
    x = int("abc")
except ValueError:
    print("Invalid conversion")

Catching specific errors gives more control and avoids hiding real bugs.

✅ Catching Multiple Exceptions :

You can use a tuple to catch multiple types in one except.

✅ Generic Exception (Not Recommended for Production) :

Only use this in debugging or very general fallback cases.

✅ else and finally Blocks :

| Block     | Runs When...                           |
| --------- | -------------------------------------- |
| `else`    | No exception is raised                 |
| `finally` | Always runs (even if exception occurs) |


✅ Raising Exceptions Manually

Used to enforce conditions or validate input.

✅ Custom Exception Classes

Custom logic for domain-specific errors.

🧾 Common Built-in Exceptions :
| Exception           | Trigger                  |
| ------------------- | ------------------------ |
| `ZeroDivisionError` | Divide by zero           |
| `ValueError`        | Invalid value for type   |
| `TypeError`         | Wrong type of operation  |
| `IndexError`        | List index out of range  |
| `KeyError`          | Missing key in dict      |
| `AttributeError`    | Invalid attribute access |
| `FileNotFoundError` | File does not exist      |
| `ImportError`       | Module not found         |
| `NameError`         | Variable not defined     |

📌 Summary Table :
| Concept             | Description                 | Example                                 |
| ------------------- | --------------------------- | --------------------------------------- |
| `open()`            | Open a file                 | `open("file.txt", "r")`                 |
| `.read(), .write()` | Read/write contents         | `file.read()`, `file.write("data")`     |
| `with open(...)`    | Safe file handling          | Closes automatically                    |
| `try-except`        | Handle expected errors      | Catch `ValueError`, `ZeroDivisionError` |
| `else`, `finally`   | Add control flows           | Cleanup and extra steps                 |
| `raise`             | Manually raise exceptions   | `raise ValueError(...)`                 |
| `CustomException`   | Define your own error types | For special cases                       |


💡 Real-Life Use Cases :
