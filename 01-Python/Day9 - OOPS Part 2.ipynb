# 📘 Special (Dunder) Methods in Python

Definition :

Dunder methods (short for “double underscore”) are special built-in methods in Python that start and end with double underscores (like __init__, __str__, __add__, etc.).

These methods are also called:
- Magic methods
- Special methods

They are used to define custom behavior for built-in operations such as:

- Object initialization
- Printing
- Operator overloading
- Length, comparison, string conversion, etc.

🧱 Commonly Used Dunder Methods :
| Method                 | Purpose                                   |
| ---------------------- | ----------------------------------------- |
| `__init__(self)`       | Constructor; initializes new objects      |
| `__str__(self)`        | Returns readable string (`print()`)       |
| `__repr__(self)`       | Returns official string representation    |
| `__len__(self)`        | Returns the length using `len()`          |
| `__add__(self, other)` | Defines behavior for `+` operator         |
| `__eq__(self, other)`  | Defines behavior for `==`                 |
| `__lt__(self, other)`  | Less than `<` comparison                  |
| `__gt__(self, other)`  | Greater than `>` comparison               |
| `__del__(self)`        | Destructor; called when object is deleted |


**🔸 1. __init__() – Constructor**

Automatically called when an object is created. 

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Kunal", 21)
print(s1.name)  # Kunal
```

**🔸 2. __str__() – Printable String Representation**

Used by print() and str() to get a human-readable version.

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student name is {self.name}"

s = Student("Kunal")
print(s)   # Student name is Kunal
```

**🔸 3. __repr__() – Developer-Friendly Representation**

Used for debugging or logging. Called when you do repr(obj) or inspect in console.

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Student('{self.name}')"

s = Student("Kunal")
print(repr(s))  # Student('Kunal')
```

📌 Tip: Use __repr__ to return unambiguous representation, like valid Python code.

**🔸 4. __len__() – Used by len()**

```python
class Basket:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)

b = Basket(['apple', 'banana'])
print(len(b))  # 2
```

**🔸 5. Operator Overloading (e.g. __add__())**

You can define how operators like +, -, == behave for your custom class.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1 + p2)   # (4, 6)
```

**🔸 6. Comparison Overloads (__eq__, __lt__, etc.)**

```python
class Student:
    def __init__(self, marks):
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks

s1 = Student(90)
s2 = Student(90)
print(s1 == s2)  # True
```

**🔸 7. __del__() – Destructor**

Called when the object is garbage collected or deleted.

```python
class Demo:
    def __del__(self):
        print("Object is being deleted")

d = Demo()
del d  # Output: Object is being deleted
```

🧠 Summary :
| Dunder Method | Used For                         |
| ------------- | -------------------------------- |
| `__init__`    | Constructor (object creation)    |
| `__str__`     | User-friendly string (`print()`) |
| `__repr__`    | Debug representation             |
| `__len__`     | Length using `len()`             |
| `__add__`     | Overload `+`                     |
| `__eq__`      | Overload `==`                    |
| `__del__`     | Object destruction               |


✅ Best Practices

- Always define __repr__ and __str__ for custom classes.
- Use dunder methods to support built-in Python operations.
- Don’t overuse them—use only where needed.

# 📘 Decorators in OOP


What is a Decorator?

A decorator is a special function in Python that modifies or enhances another function or method without changing its actual code.

In OOP, decorators like @property, @classmethod, and @staticmethod are built-in and used to define different types of methods.

🧱 Types of OOP Method Decorators :
| Decorator       | Bound To                   | Accesses?          | Use Case                                  |
| --------------- | -------------------------- | ------------------ | ----------------------------------------- |
| `@property`     | Instance                   | `self`             | Convert method into a read-only attribute |
| `@classmethod`  | Class                      | `cls`              | Affects the class as a whole              |
| `@staticmethod` | Neither class nor instance | No `self` or `cls` | Utility function inside class             |


**🔹 1. @property – Makes Method Act Like an Attribute**

✅ Purpose:

Convert a method into a read-only property. Useful for data encapsulation and getter/setter logic.

🧠 Syntax:

```python
class Person:
    def __init__(self, name):
        self._name = name  # leading underscore = private by convention

    @property
    def name(self):
        return self._name
```

✅ Usage:

```python
p = Person("Kunal")
print(p.name)   # No () required, acts like an attribute
```

🔁 Adding Setter (Optional)

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if len(value) < 3:
            raise ValueError("Name too short!")
        self._name = value
```

```python
p = Person("Raj")
p.name = "Kunal"
print(p.name)  # Kunal
```

**🔹 2. @classmethod – Bound to Class, Not Object**

✅ Purpose:

A classmethod receives the class (cls) as the first argument. It can modify class variables or create alternate constructors.

🧠 Syntax:

```python
class Student:
    school_name = "ABC School"

    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name
```

✅ Usage:

```python
Student.change_school("XYZ School")
print(Student.school_name)  # XYZ School
```

📌 Alternate Constructor Example:

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    @classmethod
    def from_string(cls, date_str):  # alternate constructor
        day, month, year = map(int, date_str.split('-'))
        return cls(day, month, year)

d = Date.from_string("30-07-2025")
print(d.day, d.month, d.year)
```

**🔹 3. @staticmethod – Utility Function Inside Class**

✅ Purpose:

A staticmethod does not take self or cls. It behaves like a regular function that belongs to a class only for organizational reasons.

🧠 Syntax:

```python
class Math:
    @staticmethod
    def add(a, b):
        return a + b
```

✅ Usage:

```python
print(Math.add(5, 3))  # 8
```

📌 Use @staticmethod when:

- The method doesn’t need class or instance context

- It just logically belongs to the class

✅ Summary Table
| Feature       | `@property`   | `@classmethod`                    | `@staticmethod`   |
| ------------- | ------------- | --------------------------------- | ----------------- |
| Takes `self`? | ✔️            | ❌                                 | ❌                 |
| Takes `cls`?  | ❌             | ✔️                                | ❌                 |
| Used for?     | Getter/Setter | Alternate constructor, class data | Utility method    |
| Accessed via? | Instance only | Class or instance                 | Class or instance |
| Typical Use   | `object.attr` | `Class.method()`                  | `Class.method()`  |


🧠 When to Use What?
- Use @property when you want attribute access with optional validation logic.

- Use @classmethod when you need to modify class state or provide alternative ways to create objects.

- Use @staticmethod for helpers that don't access class or instance variables
