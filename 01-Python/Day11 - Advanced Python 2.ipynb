# 🧪 Testing & Debugging in Python :

**🔹 1. Testing in Python**

Testing ensures your code works correctly and reliably over time. It prevents bugs during changes or new development.

✅ Types of Testing:
| Type                   | Description                                       |
| ---------------------- | ------------------------------------------------- |
| **Unit Testing**       | Tests individual components or functions.         |
| **Integration Test**   | Tests multiple components working together.       |
| **System Testing**     | Tests the complete system end-to-end.             |
| **Regression Testing** | Ensures new code doesn't break existing features. |

# 📦 unittest – Built-in Testing Framework
Python’s standard library framework based on JUnit style.

Built-in module for writing test cases.

✅ Basic Structure:

✅ Common Assertions: 
| Method                 | Checks if…               |
| ---------------------- | ------------------------ |
| `assertEqual(a, b)`    | a == b                   |
| `assertNotEqual(a, b)` | a != b                   |
| `assertTrue(x)`        | x is `True`              |
| `assertFalse(x)`       | x is `False`             |
| `assertIs(a, b)`       | a is b (identity)        |
| `assertIn(a, b)`       | a in b                   |
| `assertRaises(Error)`  | Raises an expected error |

# 🧪 pytest – Powerful Third-Party Testing Tool
✅ Why Use pytest?

No boilerplate code (no need to subclass unittest.TestCase)

- Auto-discovery of tests

- Advanced features: fixtures, parametrize, plugins

- A powerful testing framework.

✅ Basic Example:

```python
# test_math.py
def add(x, y):
    return x + y

def test_add():
    assert add(2, 3) == 5
    assert add(0, 0) == 0
```

Run: pytest test_math.py

✅ 3. assert Statement – Quick Inline Testing

- The assert keyword checks for conditions during development
- Ensures that a condition is true; raises AssertionError if not.

```python
x = 5
assert x > 0, "x should be positive"
```

- Raises AssertionError if condition is False.

- Use only during development, not production.

# 🐞 4. Debugging in Python

Debugging is the process of identifying and fixing issues in your code.

🧰 Tools for Debugging

✅ pdb – Python Debugger (Built-in)
- Interactive command-line debugger.
- Built-in interactive debugger.

🎯 Common Commands in pdb:
| Command | Description                   |
| ------- | ----------------------------- |
| `n`     | Next line                     |
| `c`     | Continue execution            |
| `l`     | List current code             |
| `s`     | Step into function            |
| `q`     | Quit debugger                 |
| `p var` | Print value of variable `var` |


Run in terminal: python -m pdb script.py

# 🧹 5. Linting – Enforcing Code Standards
✅ What is Linting?

Linting checks code for syntax/style errors, unused imports, bad variable names, etc.

📦 Tool: flake8

- Checks for PEP8 violations (Python style guide)

- Warns about unused variables, long lines, indentation

Install: pip install flake8

Run: flake8 yourfile.py

# 🎨 6. Code Formatting – Auto Format Code
📦 Tool: black

"Black is the uncompromising Python code formatter."

- Automatically formats code to PEP8

- Ensures consistent style across codebase

- Fixes indentation, line spacing, and alignment

Install: pip install black

Format file: black yourfile.py

🔧 Best Practices :

✅ Test early and often

✅ Use meaningful test names

✅ Group related tests in classes

✅ Keep test functions short

✅ Mock external APIs

✅ Run tests automatically in CI/CD

✅ Use pytest for scalable test suites

✅ Keep assert for quick sanity checks

✅ Use flake8 and black in VSCode or pre-commit hooks
