# 🌐 1. Web Development with Python

✅ Frameworks: Flask vs Django

| Feature     | Flask                      | Django                            |
| ----------- | -------------------------- | --------------------------------- |
| Type        | Micro-framework            | Full-stack framework              |
| Flexibility | More flexible, lightweight | Convention over configuration     |
| Learning    | Easier for beginners       | Steeper learning curve            |
| Use Case    | Small to medium projects   | Large, scalable apps              |
| Features    | Add what you need          | Comes with ORM, auth, admin, etc. |


🧱 Flask Basics :

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Hello Flask!"

if __name__ == '__main__':
    app.run(debug=True)
```

📦 REST APIs with Flask

```python
from flask import Flask, jsonify, request

app = Flask(__name__)

@app.route('/api/data', methods=['GET'])
def get_data():
    return jsonify({'name': 'Kunal', 'role': 'Python Dev'})
```

✅ Use JSON for API data, and tools like Postman or requests for testing.

🎨 Templating with Jinja2 (Used in Flask)

```python
from flask import render_template
@app.route("/user/<name>")
def user(name):
    return render_template("home.html", name=name)
```

✅ Features:

- {{ variable }}

- {% for item in list %}

- {% if %}...{% endif %}

🗃️ Databases in Flask

# 🖥️ 3. GUI/Desktop Development in Python
✅ tkinter – Built-in GUI Library

```python
import tkinter as tk

root = tk.Tk()
root.title("My App")
label = tk.Label(root, text="Hello GUI!")
label.pack()
root.mainloop()
```

✅ Supports:

- Buttons
- TextBoxes
- Canvas
- Frames & Menus

```python
from PyQt5.QtWidgets import QApplication, QLabel, QWidget
import sy

app = QApplication(sys.argv)
window = QWidget()
window.setWindowTitle('PyQt App')
window.setGeometry(100, 100, 300, 200)
label = QLabel('Hello PyQt', parent=window)
window.show()
sys.exit(app.exec_())
```
