# 🧠 Python Loops – Deep Notes (Basics to Advanced) :

🔹 1. What is a Loop?

A loop is a programming construct that repeats a block of code multiple times until a condition is met or an iterable is exhausted.

Python supports two main types of loops:

- for loop – iterates over sequences (like lists, strings, tuples, etc.)

- while loop – runs based on a Boolean condition


✅ How it works:
- Python fetches each item from the iterable and assigns it to the variable.

- Executes the block inside the loop.

- Repeats this until the iterable is exhausted.

✅ Example:

```python
for i in range(5):
    print("Looping:", i)
```

📍 Notes:
- range(n) gives values from 0 to n-1

- Any iterable can be looped through: strings, lists, tuples, sets, dictionaries, etc.

⚠️ Pitfall:

```python
for i in range(5):
    i = 100  # Modifying loop variable has no effect on actual sequence
```

✅ Example:

```python
x = 0
while x < 3:
    print(x)
    x += 1
```

⚠️ Warning:

Make sure the condition eventually becomes False — otherwise, you'll have an infinite loop.

🧯 Loop Control Statements :-

✅ break – exits the loop:

```python
for i in range(10):
    if i == 4:
        break
    print(i)
```

✅ continue – skips to next iteration:

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

✅ pass – does nothing (used as a placeholder):

```python
for i in range(3):
    pass
```

```python
for i in range(2):
    for j in range(3):
        print(f"i={i}, j={j}")
```

Use Case:

Useful in working with matrices, grids, or combinations.

```python
#Examples:

range(5)        # 0 to 4
range(2, 6)     # 2, 3, 4, 5
range(1, 10, 2) # 1, 3, 5, 7, 9
```

⚠️ range returns a lazy object in Python 3 — convert to list if needed.

📦 Looping through Data Structures : 


- List:

```python
for item in [1, 2, 3]:
    print(item)
```

- String:

```python
for char in "hello":
    print(char)
```

- Dictionary:

```python
data = {"name": "John", "age": 25}
for key in data:
    print(key, data[key])

for key, value in data.items():
    print(key, value)
```

- Set:

```python
for val in {1, 2, 3}:
    print(val)  # No guaranteed order
```

📊 8. enumerate() and zip() – Advanced Looping

✅ enumerate() – gives index and value:

```python
names = ["Alice", "Bob"]
for index, name in enumerate(names):
    print(index, name)
```

✅ zip() – pair elements from multiple iterables:

```python
names = ["Alice", "Bob"]
scores = [85, 90]
for name, score in zip(names, scores):
    print(f"{name} got {score}")
```

🧵 9. Loop with else:

The else block runs only if the loop completes normally (i.e., not broken with break).

Example:

```python
for i in range(3):
    print(i)
else:
    print("Loop finished without break.")
```

With break:

```python
for i in range(3):
    if i == 1:
        break
else:
    print("This won't be printed.")
```

📋 List Comprehension – Pythonic Looping :

✅ Basic:

```python
squares = [x**2 for x in range(5)]
```

✅ With condition:

```python
even = [x for x in range(10) if x % 2 == 0]
```

✅ Nested List Comprehension:

```python
matrix = [[i*j for j in range(3)] for i in range(3)]
```

🔄 Infinite Loops

```python
while True:
    print("Running forever...")
    break  # Break to avoid infinite loop
```

🧠 Loop Best Practices & Tips
- Avoid modifying a list while looping through it. Instead, use list comprehensions or loop over a copy.

- Use enumerate() and zip() to write cleaner and more Pythonic code.

- Always plan for exit conditions when using while loops.

- Prefer for loops for iterables and while loops for unknown iteration counts.

🚨 Common Mistakes :
| Mistake                                 | Fix                                           |
| --------------------------------------- | --------------------------------------------- |
| Forgetting to increment in `while` loop | Always modify variables involved in condition |
| Modifying iterable inside the loop      | Loop over a copy using `list()`               |
| Infinite loop due to `True` condition   | Use proper `break` or update condition        |
