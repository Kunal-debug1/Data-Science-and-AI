 # 🧠 Functional & Pythonic Programming in Python : 

 Functional programming emphasizes the use of functions as first-class citizens, avoiding changing state and mutable data.

**1. map() –** Apply a Function to All Items

- Definition: Applies a function to each item in an iterable (like a list) and returns a map object (which is iterable).

- Syntax: map(function, iterable)

```python
def square(n):
    return n * n

numbers = [1, 2, 3, 4]
result = map(square, numbers)
print(list(result))  # [1, 4, 9, 16]
```

✅ Key Points:

- Function must take one argument.
- Often used with lambda.

**2. filter() –** Filter Items Using a Condition

- Definition: Filters elements in an iterable for which the function returns True.
- Syntax: filter(function, iterable)

```python
def is_even(n):
    return n % 2 == 0

numbers = [1, 2, 3, 4, 5]
result = filter(is_even, numbers)
print(list(result))  # [2, 4]
```

✅ Key Points:

- Returns an iterator.
- Use with custom or lambda functions.

**3. reduce() –** Reduce a List to a Single Value

- Definition: Applies a rolling computation to sequential pairs of elements (must import from functools).
- Syntax: reduce(function, iterable)
- Requires: from functools import reduce

```python
from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)  # 10
```

✅ Key Points:

- Operates cumulatively.
- Think: summing, multiplying, finding max/min.

**4. zip() –** Combine Iterables Pairwise

- Definition: Combines multiple iterables element-wise into tuples.
- Syntax: zip(iter1, iter2, ...)

```python
a = [1, 2, 3]
b = ['a', 'b', 'c']
zipped = zip(a, b)
print(list(zipped))  # [(1, 'a'), (2, 'b'), (3, 'c')]
```

✅ Key Points:

- Stops at the shortest iterable.

**5. enumerate() –** Index + Item
   
- Definition: Adds a counter (index) to an iterable.
- Syntax: enumerate(iterable, start=0)

```python
languages = ['Python', 'Java', 'C++']
for index, lang in enumerate(languages, start=1):
    print(index, lang)
```

**6. List Comprehensions –** Pythonic & Concise
   
- Definition: A more concise and readable way to create lists.
- Syntax: [expression for item in iterable if condition]

```python
squares = [x**2 for x in range(5)]
print(squares)  # [0, 1, 4, 9, 16]
```

✅ Variants:

- With condition: [x for x in range(10) if x % 2 == 0]

- Nested loop: [x*y for x in range(1, 3) for y in range(1, 4)]

**7. Generators & yield –** Memory Efficient Iteration

✅ Generator Function

- A function that yields values using yield instead of return.
- yield: Pauses function saving its state for next call

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for num in countdown(3):
    print(num)
```

✅ Generator Expression (like list comprehension)

```python
gen = (x**2 for x in range(5))
print(next(gen))  # 0
```

✅ Benefits:

- Lazily evaluated.

- Saves memory compared to lists.

```python
nums = iter([1, 2, 3])
print(next(nums))  # 1
print(next(nums))  # 2
```

✅ Use cases:

- Behind the scenes in loops.

- Custom data streams.

**9. Decorators –** Enhance Function Behavior

Definition: A function that modifies another function.

```python
def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@my_decorator
def greet():
    print("Hello")

greet()
```

✅ Use cases:

- Logging
- Authorization
- Timing

**10. Closures –** Functions Inside Functions
    
Definition: An inner function remembers the variables of the outer function even if the outer function is done executing.

```python
def outer(msg):
    def inner():
        print(msg)
    return inner

hello = outer("Hi")
hello()  # Hi
```

✅ Useful for:

- Function factories

- Decorators

🔹 Pythonic Principles
- Use list comprehensions over loops

- Use zip(), enumerate() instead of manual indexing

- Use get() on dictionaries

- Use unpacking (a, b = b, a)

- Use generators for large data
