OOP is a programming paradigm that models real-world entities as objects, combining data (attributes) and behavior (methods).

- Promotes reusability (via inheritance)

# 🧱 Class and Object :
🔹 What is a Class?

A class is a blueprint for creating objects.
Objects are instances of classes that contain attributes (data) and methods (functions).

🔧 Defining a Class

✅ Syntax:

```python
class ClassName:
    # class body
    def __init__(self, parameters):
        self.attribute = value

    def method(self):
        # method body
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name  # instance variable
        self.age = age

    def greet(self):
        print(f"Hello, I'm {self.name} and I'm {self.age} years old.")
```

✅ Example: 

```python
class Person:
    def __init__(self, name, age):  # Constructor
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
```

🎯 Creating Objects (Instances) :1

```python
p1 = Person("Alice", 25)
p2 = Person("Bob", 30)

p1.greet()  # Hello, my name is Alice and I am 25 years old.
p2.greet()  # Hello, my name is Bob and I am 30 years old.
```

# 🔹 Encapsulation
✅ Meaning:

Protects internal object state. In Python, it's done via:

- _var : protected (convention)

- __var : private (name mangling)

✅ Example:

```python
class Account:
    def __init__(self, balance):
        self.__balance = balance  # private

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
```

# 🔹 Inheritance in Python
🧠 What is Inheritance?

Inheritance is an object-oriented programming (OOP) concept where a child class (also called subclass or derived class) can acquire the properties and behaviors (attributes and methods) of a parent class (also called superclass or base class).

It allows:
- Code reusability
- Implementation of hierarchies
- Polymorphism (overriding parent methods)

🔧 Basic Syntax

🧱 Key Terms :
| Term                              | Definition                                                              |
| --------------------------------- | ----------------------------------------------------------------------- |
| **Class**                         | A blueprint for creating objects with attributes and methods            |
| **Object**                        | A real-world instance of a class                                        |
| **Parent class (Base class)**     | A class that is being inherited from                                    |
| **Child class (Derived class)**   | A class that inherits from a parent class                               |
| **Method Overriding**             | Redefining a parent method in a child class                             |
| **super()**                       | A function to call methods from the parent class                        |
| **MRO (Method Resolution Order)** | The order in which classes are searched for a method during inheritance |

🧠 Why Use Inheritance?
- To avoid code duplication
- To improve modularity
- To represent "is-a" relationships (e.g., Dog is-a Animal)
- To extend or customize functionality

✅ Example:

```python
class Animal:                     # Parent class
    def speak(self):
        print("The animal makes a sound")

class Dog(Animal):               # Child class
    def bark(self):
        print("The dog barks")

d = Dog()
d.speak()   # Inherited from Animal
d.bark()    # Defined in Dog
```

📌 Types of Inheritance in Python :
| Type                     | Description                    |
| ------------------------ | ------------------------------ |
| Single Inheritance       | One child, one parent          |
| Multilevel Inheritance   | Inherits through a chain       |
| Multiple Inheritance     | Inherits from multiple parents |
| Hierarchical Inheritance | One parent, many children      |
| Hybrid Inheritance       | Combination of multiple types  |

# 📂 Types of Inheritance in Python

1️⃣ Single Inheritance :

A child class inherits from one parent class.

```python
class Parent:
    def show(self):
        print("Parent class")

class Child(Parent):
    def display(self):
        print("Child class")
```

2️⃣ Multilevel Inheritance :

A class inherits from a child class that itself inherited from another class.

```python
class A:
    def methodA(self):
        print("Class A")

class B(A):
    def methodB(self):
        print("Class B")

class C(B):
    def methodC(self):
        print("Class C")

obj = C()
obj.methodA()  # Inherited from A
```

3️⃣ Multiple Inheritance :

A child class inherits from multiple parent classes.

```python
class Father:
    def skills(self):
        print("Gardening, Programming")

class Mother:
    def talents(self):
        print("Singing, Painting")

class Child(Father, Mother):
    pass

c = Child()
c.skills()
c.talents()
```

4️⃣ Hierarchical Inheritance :

Multiple child classes inherit from a single parent class.

```python
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child1(Parent):
    pass

class Child2(Parent):
    pass
```

🔧 The super() Function

super() is used to call methods from the parent class, especially useful in method overriding or constructors.

✅ Example with Constructor:

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)  # Call parent constructor
        self.grade = grade
```

# 🎭 Method Overriding

When a child class defines a method with the same name as a method in the parent class, the child’s version overrides the parent’s version.

```python
class Animal:
    def speak(self):
        print("Generic animal sound")

class Cat(Animal):
    def speak(self):
        print("Meow")

c = Cat()
c.speak()  # Meow (overridden)
```

# 🧮 MRO – Method Resolution Order
MRO defines the order in which classes are searched for a method when multiple inheritance is used.

```python
class A: pass
class B(A): pass
class C(A): pass
class D(B, C): pass

print(D.__mro__)
```

# 🧠 Inheritance Best Practices
- Use inheritance only when a clear “is-a” relationship exists.
- Prefer composition over inheritance for complex logic.
- Use super() to call parent methods safely.
 -Avoid deep inheritance chains (3+ levels).

# 📘 Polymorphism in Python :

Definition of Polymorphism -

Polymorphism is an Object-Oriented Programming (OOP) concept that allows objects of different classes to be treated as objects of a common superclass.
The word “polymorphism” means “many forms.”

In Python, polymorphism allows the same method name to behave differently depending on the object calling it.

✅ Real-Life Example

Think of the action draw():

- A Circle will implement draw() to draw a circle.
- A Rectangle will implement draw() to draw a rectangle.

Despite having the same method name, the behavior varies based on the object.

🧱 Types of Polymorphism in Python :
| Type                      | Description                                                    |
| ------------------------- | -------------------------------------------------------------- |
| **Duck Typing**           | Pythonic way of implementing polymorphism using dynamic typing |
| **Method Overriding**     | Redefining a method in the child class                         |
| **Operator Overloading**  | Changing the behavior of an operator based on operands         |
| **Function Polymorphism** | Same function name with different types of input               |

**🔸 1. Duck Typing ("If it walks like a duck...")**

Python is dynamically typed, so it does not require explicit interfaces.

If an object has the required method, it will work.

```python
class Bird:
    def fly(self):
        print("Bird is flying")

class Airplane:
    def fly(self):
        print("Airplane is flying")

def let_it_fly(thing):
    thing.fly()

let_it_fly(Bird())       # Bird is flying
let_it_fly(Airplane())   # Airplane is flying
```

✔ Here, both objects have fly() method—Python doesn’t care about their types.

**🔸 2. Method Overriding (Runtime Polymorphism)**

Child class overrides a method from the parent class to give a different behavior.

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

def animal_speak(animal):
    animal.speak()

animal_speak(Dog())  # Dog barks
animal_speak(Cat())  # Cat meows
```

**🔸 3. Operator Overloading**

You can change the behavior of Python operators like +, *, ==, etc., for custom classes using dunder (magic) methods like __add__, __eq__, __str__.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1 + p2)   # (4, 6)
```

**🔸 4. Function Polymorphism** 

Functions like len(), max(), min() behave differently depending on the type of input.

```python
print(len("hello"))     # 5 (string)
print(len([1, 2, 3]))    # 3 (list)
print(len({1: "a", 2: "b"}))  # 2 (dictionary)
```

🧠 Same function → different behavior → based on data type.

🧠 Why is Polymorphism Important?
- Improves flexibility and reusability
- Enables code generalization
- Encourages abstraction
- Supports dynamic behavior

✅ Quick Revision :

```python
# Polymorphism with class methods
class Shape:
    def draw(self):
        print("Drawing a shape")

class Circle(Shape):
    def draw(self):
        print("Drawing a circle")

class Square(Shape):
    def draw(self):
        print("Drawing a square")

for shape in (Circle(), Square()):
    shape.draw()
```
