# Phase 2 – NumPy Operations

Clean examples with explanations.

```python
import numpy as np
```

## 1. Indexing & Slicing

```python
arr=np.arange(1,11)
print(arr[2:7])
print(arr[::-1])
print(arr[::2])
```

## 2. Sorting

```python
arr=np.array([5,3,1,4,2])
print(np.sort(arr))

matrix=np.array([[3,1],[1,2],[2,3]])
print(np.sort(matrix,axis=0))  # column-wise
print(np.sort(matrix,axis=1))  # row-wise
```

## 3. Filtering

```python
numbers=np.arange(1,11)
print(numbers[numbers%2==0])
print(numbers[numbers>5])
```

## 4. Fancy Indexing

```python
numbers=np.array([10,20,30,40,50])
indices=[0,2,4]
print(numbers[indices])
```

## 5. Searching

```python
numbers=np.arange(1,11)
print(np.where(numbers>5))
```

## 6. Insert & Delete

```python
arr=np.array([1,2,3,4])
print(np.insert(arr,2,99))
print(np.delete(arr,2))
```

## Common Mistakes
- Use `np.where`, not `np.wheree`.
- `np.delete()` returns a new array.
- Boolean filtering syntax: `arr[condition]`.
