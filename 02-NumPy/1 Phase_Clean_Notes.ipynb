# Phase 1 – NumPy Fundamentals

## Notes
- NumPy is a fast numerical computing library.
- Arrays are faster and more memory efficient than Python lists.
- Arrays can be 1D, 2D, or higher dimensional.


```python
import numpy as np
```

## 1. Creating Arrays

```python
arr_1d = np.array([1,2,3,4,5])
arr_2d = np.array([[1,2,3],[4,5,6]])

print(arr_1d)
print(arr_2d)
```

### Explanation
- `np.array()` converts Python lists into NumPy arrays.

## 2. Python List vs NumPy Array

```python
py_list=[1,2,3]
np_array=np.array([1,2,3])

print(py_list*2)      # repeats list
print(np_array*2)     # element-wise multiplication
```

## 3. Creating Arrays from Scratch

```python
print(np.zeros((2,3)))
print(np.ones((2,3)))
print(np.full((2,2),7))
print(np.arange(0,10,2))
print(np.random.random((2,2)))
```

## 4. Vector, Matrix, Tensor

```python
vector=np.array([1,2,3])
matrix=np.array([[1,2],[3,4]])
tensor=np.array([[[1,2],[3,4]],[[5,6],[7,8]]])

print(vector)
print(matrix)
print(tensor)
```

## 5. Array Properties

```python
arr=np.array([[1,2,3],[4,5,6]])
print("Shape:",arr.shape)
print("Dimensions:",arr.ndim)
print("Size:",arr.size)
print("Data Type:",arr.dtype)
```

## 6. Reshaping

```python
arr=np.arange(12)
print(arr.reshape(3,4))
print(arr.flatten())
```

## Key Takeaways
- Prefer NumPy for numerical work.
- Learn shape, ndim, size and reshape well.
